## 第 2 步：导入核心模块

```python
import numpy as np
import matplotlib.pyplot as plt

from pyotf.otf import HanserPSF
from pyotf.microscope import ConfocalMicroscope, FourPiConfocalMicroscope
from pyotf.vector import richards_wolf_linearly_polarized_psf

print("导入成功 ✅")
```

## 第 3 步：最小 HanserPSF 示例

```python
model = HanserPSF(wl=520, na=1.1, ni=1.33, res=80, size=64, zres=120, zsize=32)
print("PSFi shape:", model.PSFi.shape)
print("PSFi sum:", model.PSFi.sum())
```

## 第 4 步：显示中心切片

```python
z0 = model.PSFi.shape[0] // 2
plt.figure(figsize=(5,4))
plt.imshow(model.PSFi[z0], cmap="inferno")
plt.title("HanserPSF 中心 z 切片")
plt.colorbar()
plt.show()
```
